#使用cos sin就可以畫個圓

```python
%pylab inline
```

```python
theta = linspace(0,360,360)
```

```python
x = 5 * cos(theta)
```

```python
y = 5 * sin(theta)
```

```python
plot(x, y, 'ro')
axes().set_aspect('equal')
```

```python
axes().set_aspect('equal', 'datalim')
```
